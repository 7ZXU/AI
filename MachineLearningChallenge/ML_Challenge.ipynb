### 데이터 특징 추출
* 학습에 사용된 데이터 특징과 동일하게 추출 (순서도 동일)

```python
NoOfData    = 10  # Fault1, Fault2 각 10개씩 
NoOfSensor  = 4   # 가속도(Acceleration), 전압(Voltage), 전류(Current), 
NoOfFeature = 10   # 특징 개수:10개 (순서: Max, Min, Mean, RMS, Variance, Skewness, Kurtosis, Crest factor, Impulse factor, Shape factor)
def rms(x): # RMS 함수 정의
    return np.sqrt(np.mean(x**2)) #제곱평균제곱근
```

```python
FeatureData_Fault1   = np.zeros((NoOfSensor*NoOfFeature , NoOfData))
FeatureData_Fault2   = np.zeros((NoOfSensor*NoOfFeature , NoOfData))
#np.zeros 0으로된 행렬 반환
#np.zeors(행, 열) = (40,10)

for i in range(NoOfData) : 
    s1 = 'temp_data1 = TestData%d'%(i+1) #range() : 0부터 시작 #1~10
    s2 = 'temp_data2 = TestData%d'%(i+11) #11~20
    exec(s1)
    exec(s2)
    
    # Time Domain 특징값 추출
    for j in range(NoOfSensor):
        
        # Fault1 Time Domain Feature
        FeatureData_Fault1[10*j+0, i] = np.max(temp_data1.iloc[:,j])
        FeatureData_Fault1[10*j+1, i] = np.min(temp_data1.iloc[:,j])
        FeatureData_Fault1[10*j+2, i] = np.mean(temp_data1.iloc[:,j])
        FeatureData_Fault1[10*j+3, i] = rms(temp_data1.iloc[:,j])
        FeatureData_Fault1[10*j+4, i] = np.var(temp_data1.iloc[:,j])
        FeatureData_Fault1[10*j+5, i] = sp.skew(temp_data1.iloc[:,j])
        FeatureData_Fault1[10*j+6, i] = sp.kurtosis(temp_data1.iloc[:,j])
        FeatureData_Fault1[10*j+7, i] = np.max(temp_data1.iloc[:,j])/rms(temp_data1.iloc[:,j])
        FeatureData_Fault1[10*j+8, i] = rms(temp_data1.iloc[:,j])/np.mean(temp_data1.iloc[:,j])
        FeatureData_Fault1[10*j+9, i] = np.max(temp_data1.iloc[:,j])/np.mean(temp_data1.iloc[:,j])
        
        # Fault2 Time Domain Feature
        FeatureData_Fault2[10*j+0, i] = np.max(temp_data2.iloc[:,j])
        FeatureData_Fault2[10*j+1, i] = np.min(temp_data2.iloc[:,j])
        FeatureData_Fault2[10*j+2, i] = np.mean(temp_data2.iloc[:,j])
        FeatureData_Fault2[10*j+3, i] = rms(temp_data2.iloc[:,j])
        FeatureData_Fault2[10*j+4, i] = np.var(temp_data2.iloc[:,j])
        FeatureData_Fault2[10*j+5, i] = sp.skew(temp_data2.iloc[:,j])
        FeatureData_Fault2[10*j+6, i] = sp.kurtosis(temp_data2.iloc[:,j])
        FeatureData_Fault2[10*j+7, i] = np.max(temp_data2.iloc[:,j])/rms(temp_data2.iloc[:,j])
        FeatureData_Fault2[10*j+8, i] = rms(temp_data2.iloc[:,j])/np.mean(temp_data2.iloc[:,j])
        FeatureData_Fault2[10*j+9, i] = np.max(temp_data2.iloc[:,j])/np.mean(temp_data2.iloc[:,j])
        
print(FeatureData_Fault1.shape)
print(FeatureData_Fault2.shape)
```

### 데이터 특징 선택
* 주의 : 검사 데이터로 별도의 t-Test를 수행(P-value 계산)하지 말 것 >> 학습 과정에서 선택된 특징과 동일한 특징 선택 및 동일한 순서로 배열

```python
FeatureSelected_Trained = pd.DataFrame(np.concatenate([FeatureData_Fault1,FeatureData_Fault2],axis=1)) #열 기준으로 이어 붙여라 (40, 20)
FeatureData_Fault1_df = pd.DataFrame(FeatureData_Fault1)
FeatureData_Fault2_df = pd.DataFrame(FeatureData_Fault2)
```

```python
P_value_Rank = pd.read_csv('/content/drive/MyDrive/colab/BDnML/ML_ChallengeMaterials/MLmodel/P_value_Rank',header = None)
P_value_Rank
#p_value 낮은 순서로 배열 (1열 P_value)
```

```python
Rank = 5
NoOfData = 10

Fault1   = np.zeros((Rank,NoOfData))
Fault2   = np.zeros((Rank,NoOfData))

for i in range(Rank):
    #FeatureData_Fault (40,10)
    index         = int(P_value_Rank.iloc[i,0])                      # P-Value의 [0]열은 해당 통계적 특징의 행 위치를 뜻한다.
    Fault1[i,:]   = FeatureData_Fault1_df.iloc[index,:].values          # Normal각 행에 index 째의 데이터 저장(P value가 가장 낮은 순으로)
    Fault2[i,:] = FeatureData_Fault2_df.iloc[index,:].values    #(5, 10)

# 정상, 고장 특징값 합치기    
FeatureSelected = pd.DataFrame(np.transpose(np.concatenate([Fault1, Fault2] , axis=1))) #(5, 20)
#np.tranpose : 행과 열 바꾸기 #(20,5)

print("Selected Feature Data Size :", FeatureSelected.shape)
FeatureSelected
```
